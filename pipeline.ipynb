```python
import pandas as pd
import json
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

# ---------------------------
# Download NLTK resources
# ---------------------------
nltk.download("punkt")
nltk.download("stopwords")

# ---------------------------
# 1️⃣ Load JSON datasets
# ---------------------------
files = ["constitution_qa.json", "crpc_qa.json", "ipc_qa.json"]

records = []

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        records.extend(data)

# Convert to DataFrame
df = pd.DataFrame(records)

print("Dataset Loaded Successfully")
print(df.head())
print("Columns:", df.columns)

# ---------------------------
# 2️⃣ Clean the text
# ---------------------------
stop_words = set(stopwords.words("english"))

def clean_text(text):
    words = nltk.word_tokenize(str(text).lower())
    words = [w for w in words if w.isalpha() and w not in stop_words]
    return " ".join(words)

# Clean the legal question text
df["Clean_Text"] = df["question"].apply(clean_text)

print("Text Cleaning Completed")

# ---------------------------
# 3️⃣ Extract keywords using TF-IDF
# ---------------------------
vectorizer = TfidfVectorizer(max_features=10)

X = vectorizer.fit_transform(df["Clean_Text"])

keywords = vectorizer.get_feature_names_out()

print("Extracted Keywords:", keywords)

# ---------------------------
# 4️⃣ Create keyword column
# ---------------------------
df["Keywords"] = df["Clean_Text"].apply(lambda x: " ".join(x.split()[:5]))

# ---------------------------
# 5️⃣ Prepare final dataset
# ---------------------------
final_df = df[["question", "Keywords", "answer"]]

print(final_df.head())

# ---------------------------
# 6️⃣ Save processed dataset
# ---------------------------
final_df.to_csv("processed_cases.csv", index=False)

print("Processed dataset saved as processed_cases.csv")
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# Load dataset
data = pd.read_csv("processed_cases.csv")

print(data.head())

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
embeddings = model.encode(data["Clean_Text"].tolist())

print("Embeddings created:", len(embeddings))
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=5)

labels = clustering_model.fit_predict(embeddings)

data["Cluster"] = labels

data.head()
```

```python
print(data["Cluster"].value_counts())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)

print("Similarity matrix shape:", similarity_matrix.shape)

similarity_matrix[:5, :5]
```

```python
def find_similar_cases(case_index, top_n=3):
    
    similarities = similarity_matrix[case_index]
    
    similar_indices = similarities.argsort()[-top_n-1:-1][::-1]
    
    return data.iloc[similar_indices][["question","Cluster"]]
```
